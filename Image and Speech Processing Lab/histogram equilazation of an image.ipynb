```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# Load grayscale image
img = cv.imread('r ph1.jpg', 0)

# Display original image
plt.subplot(2, 2, 1)
plt.title('Original Image')
plt.imshow(img, cmap='gray')

# Display histogram of original image
plt.subplot(2, 2, 2)
plt.title('Original Histogram')
plt.hist(img.flatten(), 256, [0, 256], color='r')
plt.xlabel('Bins')

# Perform histogram equalization
img_eq = cv.equalizeHist(img)

# Display equalized image
plt.subplot(2, 2, 3)
plt.title('Equalized Image')
plt.imshow(img_eq, cmap='gray')

# Display histogram of equalized image
plt.subplot(2, 2, 4)
plt.title('Equalized Histogram')
plt.hist(img_eq.flatten(), 256, [0, 256], color='r')
plt.xlabel('Bins')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt

# Load grayscale image
img = cv.imread('r ph1.jpg', 0)

# Define the region of interest (ROI) - example: a 100x100 pixel square starting at (50, 50)
x, y, w, h = 500, 500, 1000, 1000
roi = img[y:y+h, x:x+w]

# Perform histogram equalization on the ROI
roi_eq = cv.equalizeHist(roi)

# Replace the ROI in the original image with the equalized ROI
img_eq_partial = img.copy()
img_eq_partial[y:y+h, x:x+w] = roi_eq

# Display original image with marked ROI
plt.subplot(1, 3, 1)
plt.title('Original Image with ROI')
plt.imshow(img, cmap='gray')
plt.gca().add_patch(plt.Rectangle((x, y), w, h, edgecolor='yellow', facecolor='none', linewidth=1.5))

# Display equalized ROI
plt.subplot(1, 3, 2)
plt.title('Equalized ROI')
plt.imshow(roi_eq, cmap='gray')

# Display full image with equalized ROI
plt.subplot(1, 3, 3)
plt.title('Image with Equalized ROI')
plt.imshow(img_eq_partial, cmap='gray')

plt.tight_layout()
plt.show()
```
